```python
def calcular_total(productos,iva=0.16):
  sum = 0
  for p in productos:
    sum = sum + ((p*iva) + p)

  return sum

  
```

```python
print(calcular_total([12,54,23]))
print(calcular_total(iva=0.19,productos=[12,54,23]))
```

```python
n = int(input("ingrese cantidad de números"))
for k in range(n):
  x = int(input("Ingrese " + str(k+1) + " valor:"))
  m = 0
  for i in range(x):
    if x%(i+1) == 0:
      m = m + 1
  if m > 2:
    print(x, "No es primo")
  else:
    print(x, "Es primo")
```
